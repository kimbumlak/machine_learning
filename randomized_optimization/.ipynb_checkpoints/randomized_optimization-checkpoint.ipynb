```python
def knapsack_optimization_problem():

    def random_hill_climb(problem, times, scores):
        time_start = time.time()
        state, top_score = mlrose.random_hill_climb(problem=problem)
        time_end = time.time()
        training_time = time_end - time_start
        times.append(training_time)
        scores.append(top_score)

    def simulated_annealing(problem, times, scores):
        time_start = time.time()
        state, top_score = mlrose.simulated_annealing(problem=problem)
        time_end = time.time()
        training_time = time_end - time_start
        times.append(training_time)
        scores.append(top_score)

    def genetic_alg(problem, times, scores):
        time_start = time.time()
        state, top_score = mlrose.genetic_alg(problem=problem)
        time_end = time.time()
        training_time = time_end - time_start
        times.append(training_time)
        scores.append(top_score)

    def mimic(problem, times, scores):
        time_start = time.time()
        state, top_score = mlrose.mimic(problem=problem, fast_mimic=True)
        time_end = time.time()
        training_time = time_end - time_start
        times.append(training_time)
        scores.append(top_score)

    def implement_knapsack(iter_num):
        weights = list(np.random.randint(low=1, high=100, size=iter_num))
        values = list(np.random.randint(low=1, high=100, size=iter_num))
        knapsack = mlrose.DiscreteOpt(length=iter_num, fitness_fn=mlrose.Knapsack(weights, values, 0.2))
        random_hill_climb(knapsack, hill_times, hill_scores)
        simulated_annealing(knapsack, anneal_times, anneal_scores)
        genetic_alg(knapsack, genetic_times, genetic_scores)
        mimic(knapsack, mimic_times, mimic_scores)

    hill_times, hill_scores = [], []
    anneal_times, anneal_scores = [], []
    genetic_times, genetic_scores = [], []
    mimic_times, mimic_scores = [], []    
    
    iter_range = range(10,100,20)
    for i in iter_range:
        implement_knapsack(i)

    # Plot iteration vs time
    plt.figure()
    plt.plot(iter_range, hill_times, 'o-', iter_range, anneal_times, 'o-',
             iter_range, genetic_times, 'o-', iter_range, mimic_times, 'o-')
    plt.title('Knapsack Problem Times')
    plt.xlabel('Iteration')
    plt.ylabel('Time')
    plt.legend(['Randomized Hill Climbing', 'Simulated Annealing',
                'Genetic Algorithm', 'MIMIC'])
    plt.grid()
    plt.show()

    # Plot iteration vs Fitness Value
    plt.plot(iter_range, hill_scores, 'o-', iter_range, anneal_scores, 'o-',
             iter_range, genetic_scores, 'o-', iter_range, mimic_scores, 'o-')
    plt.title('Knapsack Problem Fitness')
    plt.xlabel('Iteration')
    plt.ylabel('Fitness Value')
    plt.legend(['Randomized Hill Climbing', 'Simulated Annealing',
                'Genetic Algorithm', 'MIMIC'])
    plt.grid()
    plt.show()
```

```python
def continuous_peaks_optimization_problem():

    def random_hill_climb(problem, times, scores):
        time_start = time.time()
        state, top_score = mlrose.random_hill_climb(problem=problem)
        time_end = time.time()
        training_time = time_end - time_start
        times.append(training_time)
        scores.append(top_score)

    def simulated_annealing(problem, times, scores):
        time_start = time.time()
        state, top_score = mlrose.simulated_annealing(problem=problem)
        time_end = time.time()
        training_time = time_end - time_start
        times.append(training_time)
        scores.append(top_score)

    def genetic_alg(problem, times, scores):
        time_start = time.time()
        state, top_score = mlrose.genetic_alg(problem=problem)
        time_end = time.time()
        training_time = time_end - time_start
        times.append(training_time)
        scores.append(top_score)

    def mimic(problem, times, scores):
        time_start = time.time()
        state, top_score = mlrose.mimic(problem=problem)
        time_end = time.time()
        training_time = time_end - time_start
        times.append(training_time)
        scores.append(top_score)

    def implement_continuous_peaks(iter_num):
        continuous_peaks = mlrose.DiscreteOpt(length=iter_num, fitness_fn=mlrose.ContinuousPeaks(t_pct=0.15))
        random_hill_climb(continuous_peaks, hill_times, hill_scores)
        simulated_annealing(continuous_peaks, anneal_times, anneal_scores)
        genetic_alg(continuous_peaks, genetic_times, genetic_scores)
        mimic(continuous_peaks, mimic_times, mimic_scores)

    hill_times, hill_scores = [], []
    anneal_times, anneal_scores = [], []
    genetic_times, genetic_scores = [], []
    mimic_times, mimic_scores = [], []    
    
    iter_range = range(10,100,10)
    for i in iter_range:
        implement_continuous_peaks(i)

    # Plot iteration vs time
    plt.figure()
    plt.plot(iter_range, hill_times, 'o-', iter_range, anneal_times, 'o-',
             iter_range, genetic_times, 'o-', iter_range, mimic_times, 'o-')
    plt.title('Continuous Peak Problem Times')
    plt.xlabel('Iteration')
    plt.ylabel('Time')
    plt.legend(['Randomized Hill Climbing', 'Simulated Annealing',
                'Genetic Algorithm', 'MIMIC'])
    plt.grid()
    plt.show()

    # Plot iteration vs Fitness Value
    plt.plot(iter_range, hill_scores, 'o-', iter_range, anneal_scores, 'o-',
             iter_range, genetic_scores, 'o-', iter_range, mimic_scores, 'o-')
    plt.title('Continuous Peak Problem Fitness')
    plt.xlabel('Iteration')
    plt.ylabel('Fitness Value')
    plt.legend(['Randomized Hill Climbing', 'Simulated Annealing',
                'Genetic Algorithm', 'MIMIC'])
    plt.grid()
    plt.show()
```

```python
def flip_flop_optimization_problem():

    def random_hill_climb(problem, times, scores):
        time_start = time.time()
        state, top_score = mlrose.random_hill_climb(problem=problem)
        time_end = time.time()
        training_time = time_end - time_start
        times.append(training_time)
        scores.append(top_score)

    def simulated_annealing(problem, times, scores):
        time_start = time.time()
        state, top_score = mlrose.simulated_annealing(problem=problem)
        time_end = time.time()
        training_time = time_end - time_start
        times.append(training_time)
        scores.append(top_score)

    def genetic_alg(problem, times, scores):
        time_start = time.time()
        state, top_score = mlrose.genetic_alg(problem=problem)
        time_end = time.time()
        training_time = time_end - time_start
        times.append(training_time)
        scores.append(top_score)

    def mimic(problem, times, scores):
        time_start = time.time()
        state, top_score = mlrose.mimic(problem=problem)
        time_end = time.time()
        training_time = time_end - time_start
        times.append(training_time)
        scores.append(top_score)

    def implement_flip_flop(iter_num):
        flip_flop = mlrose.DiscreteOpt(length=iter_num, fitness_fn=mlrose.FlipFlop(), max_val=2)
        random_hill_climb(flip_flop, hill_times, hill_scores)
        simulated_annealing(flip_flop, anneal_times, anneal_scores)
        genetic_alg(flip_flop, genetic_times, genetic_scores)
        mimic(flip_flop, mimic_times, mimic_scores)

    hill_times, hill_scores = [], []
    anneal_times, anneal_scores = [], []
    genetic_times, genetic_scores = [], []
    mimic_times, mimic_scores = [], []    
    
    iter_range = range(10,100,10)
    for i in iter_range:
        implement_flip_flop(i)

    # Plot iteration vs time
    plt.figure()
    plt.plot(iter_range, hill_times, 'o-', iter_range, anneal_times, 'o-',
             iter_range, genetic_times, 'o-', iter_range, mimic_times, 'o-')
    plt.title('Flip Flop Problem Times')
    plt.xlabel('Iteration')
    plt.ylabel('Time')
    plt.legend(['Randomized Hill Climbing', 'Simulated Annealing',
                'Genetic Algorithm', 'MIMIC'])
    plt.grid()
    plt.show()

    # Plot iteration vs Fitness Value
    plt.plot(iter_range, hill_scores, 'o-', iter_range, anneal_scores, 'o-',
             iter_range, genetic_scores, 'o-', iter_range, mimic_scores, 'o-')
    plt.title('Flip Flop Problem Fitness')
    plt.xlabel('Iteration')
    plt.ylabel('Fitness Value')
    plt.legend(['Randomized Hill Climbing', 'Simulated Annealing',
                'Genetic Algorithm', 'MIMIC'])
    plt.grid()
    plt.show()
```

```python
def pre_process_on_campus_placement_data():
    # Read on campus placement records data.
    college_df = pd.read_csv('collegePlace.csv')

    # Encode string data types to integer (numerical) data type
    college_df['Gender_n'] = LabelEncoder().fit_transform(college_df['Gender'])
    college_df['Stream_n'] = LabelEncoder().fit_transform(college_df['Stream'])

    # Drop columns used to encode
    college_df_final = college_df.drop(['Gender', 'Stream'], axis='columns')

    # Split dataset into two parts: Train dataset and test dataset.
    # Features
    features_college = college_df_final.drop(columns='PlacedOrNot')

    # Target
    target_college = college_df_final['PlacedOrNot']
    return [features_college, target_college]
```

```python
def neural_network_random_hill_climb(features_train, features_test, target_train, target_test):
    iters_range = range(1, 2200, 200)
    times, accuracies = [], []
    for iters in iters_range:
        neural_network_hill = mlrose.NeuralNetwork(hidden_nodes=[3], activation='tanh',
                                                   algorithm='random_hill_climb', max_iters=iters,
                                                   learning_rate=0.1, clip_max=5, max_attempts=3, random_state=3)
        time_start = time.time()
        neural_network_hill.fit(features_train, target_train)
        time_end = time.time()
        training_time = time_end - time_start
        target_test_predict = neural_network_hill.predict(features_test)
        accuracy_score = metrics.accuracy_score(target_test, target_test_predict)
        times.append(training_time)
        accuracies.append(accuracy_score)
    return times, accuracies

def neural_network_simulated_annealing(features_train, features_test, target_train, target_test):
    iters_range = range(1, 2200, 200)
    times, accuracies = [], []
    for iters in iters_range:
        neural_network_anneal = mlrose.NeuralNetwork(hidden_nodes=[3], activation='tanh',
                                                     algorithm='simulated_annealing', max_iters=iters,
                                                     learning_rate=0.1, clip_max=5, max_attempts=3, random_state=3)
        time_start = time.time()
        neural_network_anneal.fit(features_train, target_train)
        time_end = time.time()
        training_time = time_end - time_start
        target_test_predict = neural_network_anneal.predict(features_test)
        accuracy_score = metrics.accuracy_score(target_test, target_test_predict)
        times.append(training_time)
        accuracies.append(accuracy_score)
    return times, accuracies
        
def neural_network_genetic_algorithm(features_train, features_test, target_train, target_test):
    iters_range = range(1, 2200, 200)
    times, accuracies = [], []
    for iters in iters_range:
        neural_network_genetic = mlrose.NeuralNetwork(hidden_nodes=[3], activation='tanh',
                                                      algorithm='genetic_alg', max_iters=iters,
                                                      learning_rate=0.1, clip_max=5, max_attempts=3, random_state=3)
        time_start = time.time()
        neural_network_genetic.fit(features_train, target_train)
        time_end = time.time()
        training_time = time_end - time_start
        target_test_predict = neural_network_genetic.predict(features_test)
        accuracy_score = metrics.accuracy_score(target_test, target_test_predict)
        times.append(training_time)
        accuracies.append(accuracy_score)
    return times, accuracies

def plot_randomized_algorithms(times, accuracies, title):
    iters_range = range(1, 2200, 200)
    plt.figure()
    plt.plot(iters_range, times,'o-')
    plt.title('{0} Time vs Iterations'.format(title))
    plt.xlabel('Iterations')
    plt.ylabel('Time')
    plt.grid()
    plt.show()

    plt.plot(iters_range, accuracies,'o-')
    plt.title('{0} Accuracy vs Iterations'.format(title))
    plt.xlabel('Iterations')
    plt.ylabel('Accuracy')
    plt.grid()
    plt.show()
```

```python
features_college, target_college = pre_process_on_campus_placement_data()
features_train, features_test, target_train, target_test = train_test_split(features_college, target_college,
                                                                            test_size=0.3, random_state=42)
```

```python
hill_times, hill_accuracies = neural_network_random_hill_climb(features_train, features_test,
                                                               target_train, target_test)
plot_randomized_algorithms(hill_times, hill_accuracies, 'Randomized Hill Climbing')
```

```python
anneal_times, anneal_accuracies = neural_network_simulated_annealing(features_train, features_test,
                                                                     target_train, target_test)
plot_randomized_algorithms(anneal_times, anneal_accuracies, 'Simulated Annealing')
```

```python
genetic_times, genetic_accuracies = neural_network_genetic_algorithm(features_train, features_test,
                                                                     target_train, target_test)
plot_randomized_algorithms(genetic_times, genetic_accuracies, 'Genetic Algorithm')
```

```python
def compare_accuracy_score(optimization_algorithm_lists, accuracy_lists):
    plt.figure()
    plt.barh(optimization_algorithm_lists, accuracy_lists)
    plt.title('Compare Accuracy Score')
    plt.xlabel('Accuracy')
    plt.show()
```

```python
def compare_iteration_time(optimization_algorithm_lists, iteration_time_lists):
    plt.figure()
    plt.barh(optimization_algorithm_lists, iteration_time_lists)
    plt.title('Compare Training Time')
    plt.xlabel('Training Time in seconds')
    plt.show()
```

```python
optimization_algorithm_lists = ['Randomized Hill Climbing', 'Simulated Annealing', 'Genetic Algorithm']
accuracy_lists = [max(hill_accuracies), max(anneal_accuracies), max(genetic_accuracies)]
iteration_time_lists = [max(hill_times), max(anneal_times), max(genetic_times)]
compare_accuracy_score(optimization_algorithm_lists, accuracy_lists)
compare_iteration_time(optimization_algorithm_lists, iteration_time_lists)
```
