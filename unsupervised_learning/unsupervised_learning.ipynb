```python
import pandas as pd
import numpy as np
import time
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA, FastICA, TruncatedSVD
from sklearn import random_projection
from scipy.stats import kurtosis
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
```

```python
# Pre-Processing steps for heart disease data.

def pre_process_heart_disease_data():
    # Read heart disease data.
    heart_df = pd.read_csv('heart.csv')

    # Encode string data types to integer (numerical) data type
    heart_df['Sex_n'] = LabelEncoder().fit_transform(heart_df['Sex'])
    heart_df['ChestPainType_n'] = LabelEncoder().fit_transform(heart_df['ChestPainType'])
    heart_df['RestingECG_n'] = LabelEncoder().fit_transform(heart_df['RestingECG'])
    heart_df['ExerciseAngina_n'] = LabelEncoder().fit_transform(heart_df['ExerciseAngina'])
    heart_df['ST_Slope_n'] = LabelEncoder().fit_transform(heart_df['ST_Slope'])

    # Drop columns used to encode
    heart_df_final = heart_df.drop(['Sex', 'ChestPainType', 'RestingECG', 'ExerciseAngina', 'ST_Slope'], axis='columns')
    return heart_df_final
```

```python
# Pre-Processing steps for on campus placement data.

def pre_process_on_campus_placement_data():
    # Read on campus placement records data.
    college_df = pd.read_csv('collegePlace.csv')

    # Encode string data types to integer (numerical) data type
    college_df['Gender_n'] = LabelEncoder().fit_transform(college_df['Gender'])
    college_df['Stream_n'] = LabelEncoder().fit_transform(college_df['Stream'])

    # Drop columns used to encode
    college_df_final = college_df.drop(['Gender', 'Stream'], axis='columns')
    return college_df_final
```

```python
def optimal_cluster_num_plot(data, plot_title):
    num_clusters = range(2, 15)
    sse = []
    for k in num_clusters:
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(data)
        sse.append(kmeans.inertia_)
    plt.figure()
    plt.plot(num_clusters, sse, marker='o')
    plt.title(plot_title)
    plt.xlabel('Number of Clusters')
    plt.ylabel('Distance')
    plt.grid()
    plt.show()
```

```python
def silhouette_analysis_plot(data, cluster_num_range, model, plot_title):
    silhouette_avg = []
    for num_clusters in cluster_num_range:
        clf = model(num_clusters, random_state=42)
        clf.fit(data)
        #cluster_labels = clf.labels_
        labels=clf.predict(data)
        silhouette_avg.append(silhouette_score(data, labels))
    plt.figure()
    plt.plot(cluster_num_range, silhouette_avg, marker='o')
    plt.xlabel('Number of Clusters') 
    plt.ylabel('Silhouette score') 
    plt.title(plot_title)
    plt.grid()
    plt.show()
    print(max(silhouette_avg))
```

```python
def expectation_maximization_bic_plot(data, plot_title):
    # Set up a range of cluster numbers to try
    n_range = range(1,15)
    bic_score = []
    aic_score = []
    for n in n_range:
        gm = GaussianMixture(n_components=n, random_state=42)
        gm.fit(data)
        bic_score.append(gm.bic(data))
        aic_score.append(gm.aic(data))
    plt.figure()
    plt.plot(n_range, bic_score, '-o', color='orange', label='BIC')
    plt.plot(n_range, aic_score, '-o', color='green', label='AIC')
    plt.xlabel('Number of Clusters')
    plt.ylabel('Score')
    plt.title(plot_title)
    plt.legend(loc='best')
    plt.grid()
    plt.show()
```

```python
def pca_variance_plot(data, feature_num, plot_title):
    covar_matrix = PCA(n_components=feature_num, random_state=42)
    covar_matrix.fit(data)
    variance = covar_matrix.explained_variance_ratio_
    var = np.cumsum(np.round(covar_matrix.explained_variance_ratio_, decimals=3)*100)
    plt.figure()
    plt.plot(var, marker='o')
    plt.ylabel('Variance %')
    plt.xlabel('Number of Components')
    plt.title(plot_title)
    plt.grid()
    plt.show()
```

```python
def ica_kurtosis(data, feature_num, plot_title):
    kurt_total = []
    component_num = range(2, feature_num+1)
    for num in component_num:
        ica = FastICA(n_components=num, max_iter=5000, random_state=42)
        ica.fit(data)
        kurt_comp = np.abs(kurtosis(ica.components_)).mean()
        kurt_total.append(kurt_comp)
    plt.figure()
    plt.plot(component_num, kurt_total, marker='o')
    plt.ylabel('Average Kurtosis')
    plt.xlabel('Number of Components')
    plt.title(plot_title)
    plt.grid()
    plt.show()
```

```python
def randomized_projection(data, feature_num, plot_title):
    mse = []
    component_num = range(2, feature_num+1)
    for num in component_num:
        rp = random_projection.SparseRandomProjection(n_components=num, random_state=42)
        rp_fit = rp.fit_transform(data)
        # Compute Pseudo inverse, reconstruct data and compute MSE
        P_inv = np.linalg.pinv(rp.components_.toarray())  # dimensions check : k x m -> m x k
        x_reconstructed = (P_inv @ rp_fit.T).T  # dimensions check: m x k x k x n = m x n -> n x m
        mse.append(np.mean((x_reconstructed)))
    plt.figure()
    plt.plot(component_num, mse, marker='o')
    plt.ylabel('Reconstruction Error')
    plt.xlabel('Number of Components')
    plt.title(plot_title)
    plt.grid()
    plt.show()
```

```python
def svd_variance_plot(data, feature_num, plot_title):
    covar_matrix = TruncatedSVD(n_components=feature_num, random_state=42)
    covar_matrix.fit(data)
    variance = covar_matrix.explained_variance_ratio_
    var = np.cumsum(np.round(covar_matrix.explained_variance_ratio_, decimals=3)*100)
    plt.figure()
    plt.plot(var, marker='o')
    plt.ylabel('Variance %')
    plt.xlabel('Number of Components')
    plt.title(plot_title)
    plt.grid()
    plt.show()
```

```python
def dimensional_reduction_and_clustering(data, pca_comp, ica_comp, rp_comp, svd_comp, title):
    cluster_range = [2, 3, 4, 5, 6, 7, 8]
    pca = PCA(n_components=pca_comp, random_state=42).fit(data)
    pca = pca.transform(data)
    silhouette_analysis_plot(pca, cluster_range, KMeans, title+' PCA Dimensional Reduction and Kmeans clustering')
    silhouette_analysis_plot(pca, cluster_range, GaussianMixture, title+' PCA Dimensional Reduction and Expectation Maximization clustering')
    
    ica = FastICA(n_components=ica_comp, random_state=42).fit(data)
    ica = ica.transform(data)
    silhouette_analysis_plot(ica, cluster_range, KMeans, title+' ICA Dimensional Reduction and Kmeans clustering')
    silhouette_analysis_plot(ica, cluster_range, GaussianMixture, title+' ICA Dimensional Reduction and Expectation Maximization clustering')
    
    rp = random_projection.SparseRandomProjection(n_components=rp_comp, random_state=42).fit(data)
    rp = rp.transform(data)
    silhouette_analysis_plot(rp, cluster_range, KMeans, title+' RP Dimensional Reduction and Kmeans clustering')
    silhouette_analysis_plot(rp, cluster_range, GaussianMixture, title+' RP Dimensional Reduction and Expectation Maximization clustering')
    
    svd = TruncatedSVD(n_components=svd_comp, random_state=42).fit(data)
    svd = svd.transform(data)
    silhouette_analysis_plot(svd, cluster_range, KMeans, title+' SVD Dimensional Reduction and Kmeans clustering')
    silhouette_analysis_plot(svd, cluster_range, GaussianMixture, title+' SVD Dimensional Reduction and Expectation Maximization clustering')
```

```python
def neural_network_dimensional_reduction(data):
    x = data.drop(columns='HeartDisease')
    y = data['HeartDisease']
    
    def nn_pca(x, y):
        pca = PCA(n_components=2, random_state=42).fit(x)
        X_pca = pca.transform(x)

        clf = MLPClassifier(max_iter=1000)
        X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.3, random_state=42)

        start = time.time()
        clf.fit(X_train, y_train)
        train_time = time.time() - start
        print("NN PCA Train time: " + str(train_time))

        y_pred = clf.predict(X_test)

        score = metrics.accuracy_score(y_test, y_pred)
        print("NN PCA Test Accuracy: " + str(score))
        return [train_time, score]
    
    def nn_ica(x, y):
        ica = FastICA(n_components=2, random_state=42).fit(x)
        X_ica = ica.transform(x)

        clf = MLPClassifier(max_iter=1000)
        X_train, X_test, y_train, y_test = train_test_split(X_ica, y, test_size=0.3, random_state=42)

        start = time.time()
        clf.fit(X_train, y_train)
        train_time = time.time() - start
        print("NN ICA Train time: " + str(train_time))

        y_pred = clf.predict(X_test)

        score = metrics.accuracy_score(y_test, y_pred)
        print("NN ICA Test Accuracy: " + str(score))
        return [train_time, score]
    
    def nn_rp(x, y):
        rp = random_projection.SparseRandomProjection(n_components=2, random_state=42).fit(x)
        X_rp = rp.transform(x)

        clf = MLPClassifier(max_iter=1000)
        X_train, X_test, y_train, y_test = train_test_split(X_rp, y, test_size=0.3, random_state=42)

        start = time.time()
        clf.fit(X_train, y_train)
        train_time = time.time() - start
        print("NN RP Train time: " + str(train_time))

        y_pred = clf.predict(X_test)

        score = metrics.accuracy_score(y_test, y_pred)
        print("NN RP Test Accuracy: " + str(score))
        return [train_time, score]
    
    def nn_svd(x, y):
        svd = TruncatedSVD(n_components=2, random_state=42).fit(x)
        X_svd = svd.transform(x)

        clf = MLPClassifier(max_iter=1000)
        X_train, X_test, y_train, y_test = train_test_split(X_svd, y, test_size=0.3, random_state=42)

        start = time.time()
        clf.fit(X_train, y_train)
        train_time = time.time() - start
        print("NN SVD Train time: " + str(train_time))

        y_pred = clf.predict(X_test)

        score = metrics.accuracy_score(y_test, y_pred)
        print("NN SVD Test Accuracy: " + str(score))
        return [train_time, score]
    
    pca_time, pca_score = nn_pca(x, y)
    ica_time, ica_score = nn_ica(x, y)
    rp_time, rp_score = nn_rp(x, y)
    svd_time, svd_score = nn_svd(x, y)

    return [pca_time, pca_score, ica_time, ica_score, rp_time, rp_score, svd_time, svd_score]
```

```python
def neural_network_dimensional_reduction_and_clustering(data):
    x = data.drop(columns='HeartDisease')
    y = data['HeartDisease']
    
    def nn_pca_kmean(x, y):
        pca = PCA(n_components=2, random_state=42).fit(x)
        X_pca = pca.transform(x)
        
        kmeans=KMeans(n_clusters=2, random_state=42).fit(X_pca)
        kmeans.predict(X_pca)
        y_tmp = kmeans.predict(X_pca)
        result = pd.concat([pd.DataFrame(X_pca), pd.DataFrame(y_tmp)], axis=1, sort=False)
        result.columns = [0, 1, 2]

        clf = MLPClassifier(max_iter=1000)
        X_train, X_test, y_train, y_test = train_test_split(result, y, test_size=0.3, random_state=42)

        start = time.time()
        clf.fit(X_train, y_train)
        train_time = time.time() - start
        print("NN PCA Kmean Train time: " + str(train_time))

        y_pred = clf.predict(X_test)

        score = metrics.accuracy_score(y_test, y_pred)
        print("NN PCA Kmean Test Accuracy: " + str(score))
        return [train_time, score]
    
    def nn_ica_kmean(x, y):
        ica = FastICA(n_components=2, random_state=42).fit(x)
        X_ica = ica.transform(x)
        
        kmeans=KMeans(n_clusters=2, random_state=42).fit(X_ica)
        kmeans.predict(X_ica)
        y_tmp = kmeans.predict(X_ica)
        result = pd.concat([pd.DataFrame(X_ica), pd.DataFrame(y_tmp)], axis=1, sort=False)
        result.columns = [0, 1, 2]

        clf = MLPClassifier(max_iter=1000)
        X_train, X_test, y_train, y_test = train_test_split(result, y, test_size=0.3, random_state=42)

        start = time.time()
        clf.fit(X_train, y_train)
        train_time = time.time() - start
        print("NN ICA Kmean Train time: " + str(train_time))

        y_pred = clf.predict(X_test)

        score = metrics.accuracy_score(y_test, y_pred)
        print("NN ICA Kmean Test Accuracy: " + str(score))
        return [train_time, score]
    
    def nn_rp_kmean(x, y):
        rp = random_projection.SparseRandomProjection(n_components=2, random_state=42).fit(x)
        X_rp = rp.transform(x)
        
        kmeans=KMeans(n_clusters=2, random_state=42).fit(X_rp)
        kmeans.predict(X_rp)
        y_tmp = kmeans.predict(X_rp)
        result = pd.concat([pd.DataFrame(X_rp), pd.DataFrame(y_tmp)], axis=1, sort=False)
        result.columns = [0, 1, 2]

        clf = MLPClassifier(max_iter=1000)
        X_train, X_test, y_train, y_test = train_test_split(result, y, test_size=0.3, random_state=42)

        start = time.time()
        clf.fit(X_train, y_train)
        train_time = time.time() - start
        print("NN RP Kmean Train time: " + str(train_time))

        y_pred = clf.predict(X_test)

        score = metrics.accuracy_score(y_test, y_pred)
        print("NN RP Kmean Test Accuracy: " + str(score))
        return [train_time, score]
    
    def nn_svd_kmean(x, y):
        svd = TruncatedSVD(n_components=2, random_state=42).fit(x)
        X_svd = svd.transform(x)
        
        kmeans=KMeans(n_clusters=2, random_state=42).fit(X_svd)
        kmeans.predict(X_svd)
        y_tmp = kmeans.predict(X_svd)
        result = pd.concat([pd.DataFrame(X_svd), pd.DataFrame(y_tmp)], axis=1, sort=False)
        result.columns = [0, 1, 2]

        clf = MLPClassifier(max_iter=1000)
        X_train, X_test, y_train, y_test = train_test_split(result, y, test_size=0.3, random_state=42)

        start = time.time()
        clf.fit(X_train, y_train)
        train_time = time.time() - start
        print("NN SVD Kmean Train time: " + str(train_time))

        y_pred = clf.predict(X_test)

        score = metrics.accuracy_score(y_test, y_pred)
        print("NN SVD Kmean Test Accuracy: " + str(score))
        return [train_time, score]
    
    def nn_pca_em(x, y):
        pca = PCA(n_components=2, random_state=42).fit(x)
        X_pca = pca.transform(x)
        
        em=GaussianMixture(n_components=2, random_state=42).fit(X_pca)
        em.predict(X_pca)
        y_tmp = em.predict(X_pca)
        result = pd.concat([pd.DataFrame(X_pca), pd.DataFrame(y_tmp)], axis=1, sort=False)
        result.columns = [0, 1, 2]

        clf = MLPClassifier(max_iter=1000)
        X_train, X_test, y_train, y_test = train_test_split(result, y, test_size=0.3, random_state=42)

        start = time.time()
        clf.fit(X_train, y_train)
        train_time = time.time() - start
        print("NN PCA EM Train time: " + str(train_time))

        y_pred = clf.predict(X_test)

        score = metrics.accuracy_score(y_test, y_pred)
        print("NN PCA EM Test Accuracy: " + str(score))
        return [train_time, score]
    
    def nn_ica_em(x, y):
        ica = FastICA(n_components=2, random_state=42).fit(x)
        X_ica = ica.transform(x)
        
        em=GaussianMixture(n_components=2, random_state=42).fit(X_ica)
        em.predict(X_ica)
        y_tmp = em.predict(X_ica)
        result = pd.concat([pd.DataFrame(X_ica), pd.DataFrame(y_tmp)], axis=1, sort=False)
        result.columns = [0, 1, 2]

        clf = MLPClassifier(max_iter=1000)
        X_train, X_test, y_train, y_test = train_test_split(result, y, test_size=0.3, random_state=42)

        start = time.time()
        clf.fit(X_train, y_train)
        train_time = time.time() - start
        print("NN ICA EM Train time: " + str(train_time))

        y_pred = clf.predict(X_test)

        score = metrics.accuracy_score(y_test, y_pred)
        print("NN ICA EM Test Accuracy: " + str(score))
        return [train_time, score]
    
    def nn_rp_em(x, y):
        rp = random_projection.SparseRandomProjection(n_components=2, random_state=42).fit(x)
        X_rp = rp.transform(x)
        
        em=GaussianMixture(n_components=2, random_state=42).fit(X_rp)
        em.predict(X_rp)
        y_tmp = em.predict(X_rp)
        result = pd.concat([pd.DataFrame(X_rp), pd.DataFrame(y_tmp)], axis=1, sort=False)
        result.columns = [0, 1, 2]

        clf = MLPClassifier(max_iter=1000)
        X_train, X_test, y_train, y_test = train_test_split(result, y, test_size=0.3, random_state=42)

        start = time.time()
        clf.fit(X_train, y_train)
        train_time = time.time() - start
        print("NN RP EM Train time: " + str(train_time))

        y_pred = clf.predict(X_test)

        score = metrics.accuracy_score(y_test, y_pred)
        print("NN RP EM Test Accuracy: " + str(score))
        return [train_time, score]
    
    def nn_svd_em(x, y):
        svd = TruncatedSVD(n_components=2, random_state=42).fit(x)
        X_svd = svd.transform(x)
        
        em=GaussianMixture(n_components=2, random_state=42).fit(X_svd)
        em.predict(X_svd)
        y_tmp = em.predict(X_svd)
        result = pd.concat([pd.DataFrame(X_svd), pd.DataFrame(y_tmp)], axis=1, sort=False)
        result.columns = [0, 1, 2]

        clf = MLPClassifier(max_iter=1000)
        X_train, X_test, y_train, y_test = train_test_split(result, y, test_size=0.3, random_state=42)

        start = time.time()
        clf.fit(X_train, y_train)
        train_time = time.time() - start
        print("NN SVD EM Train time: " + str(train_time))

        y_pred = clf.predict(X_test)

        score = metrics.accuracy_score(y_test, y_pred)
        print("NN SVD EM Test Accuracy: " + str(score))
        return [train_time, score]
    
    pca_kmean_time, pca_kmean_score = nn_pca_kmean(x, y)
    ica_kmean_time, ica_kmean_score = nn_ica_kmean(x, y)
    rp_kmean_time, rp_kmean_score = nn_rp_kmean(x, y)
    svd_kmean_time, svd_kmean_score = nn_svd_kmean(x, y)
    
    pca_em_time, pca_em_score = nn_pca_em(x, y)
    ica_em_time, ica_em_score = nn_ica_em(x, y)
    rp_em_time, rp_em_score = nn_rp_em(x, y)
    svd_em_time, svd_em_score = nn_svd_em(x, y)

    return [pca_kmean_time, pca_kmean_score, ica_kmean_time, ica_kmean_score,
            rp_kmean_time, rp_kmean_score, svd_kmean_time, svd_kmean_score,
            pca_em_time, pca_em_score, ica_em_time, ica_em_score,
            rp_em_time, rp_em_score, svd_em_time, svd_em_score]
```

```python
def compare_accuracy_score(classifier_lists, accuracy_lists, plot_title):
    plt.figure()
    plt.barh(classifier_lists, accuracy_lists)
    plt.title(plot_title)
    plt.xlabel('Accuracy')
    plt.show()
```

```python
def compare_training_time(classifier_lists, training_time_lists, plot_title):
    plt.figure()
    plt.barh(classifier_lists, training_time_lists)
    plt.title(plot_title)
    plt.xlabel('Training Time in seconds')
    plt.show()
```
